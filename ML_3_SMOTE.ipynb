## SMOTE(Synthetic Minority Oversampling Technique)
SMOTE (Synthetic Minority Over-sampling Technique) is a technique used in machine learning to address imbalanced datasets where the minority class has significantly fewer instances than the majority class. SMOTE involves generating synthetic instances of the minority class by interpolating between existing instances.

#### It is an interpolation technique.
#### Interpolation Technique: In interpolation technique,we generate the new points instead of repeating the same points.

```python
from sklearn.datasets import make_classification
```

```python
## X independent feature
## y dependent feature
X,y = make_classification(n_samples=1000, n_features=2, n_redundant=0, n_clusters_per_class=1, weights=[0.90],random_state=1)
```

```python
import pandas as pd
```

```python
#Creating a dataset containing f1,f2,target
## Converting it into Dataframe

df1 = pd.DataFrame(X,columns = ['f1','f2'])
df2 = pd.DataFrame(y,columns= ['target'])
final_df = pd.concat([df1,df2],axis = 1)
```

```python
final_df
```

```python
final_df['target'].value_counts()
```

#### Imbalacned dataset created.

```python
import matplotlib.pyplot as plt
```

```python
#Plotting the dataset

plt.scatter(final_df['f1'] , final_df['f2'], c= final_df['target'] )
```

### Now applying the SMOTE technique to generate new points for minority data points:

```python
pip install imblearn # installing library ibm(imbalance) learn
```

```python
from imblearn.over_sampling import SMOTE
```

```python
## TRANSFORM THE DATASET 

oversample = SMOTE()
X,y = oversample.fit_resample(final_df[['f1','f2']], final_df['target'])
```

```python
X.shape ## 1788 rows 2 column(f1, f2)
```

```python
y.shape ## 1788 rows, 1 column(target)
```

```python
## Checking the number of the datapoints
len(y[y==1])
```

```python
len(y[y==0])
```
